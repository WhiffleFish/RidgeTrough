```python
from netCDF4 import Dataset
import datetime

def year_range(hgt_path):
    files = os.listdir(HGT_PATH)
    min_year = float('inf')
    max_year =  float('-inf')
    for file in files:
        m = re.search(hgt_re, file)
        if m is not None:
            year = int(m.group(0).split('.')[1])
            if year < min_year:
                min_year = year
            if year > max_year:
                max_year = year
    
    return min_year, max_year

def retrieve_data(path):
    fh = Dataset(path, mode='r')
    lons = fh.variables['lon'][:].data
    lats = fh.variables['lat'][:].data
    times = fh.variables['time'][:].data
    times = times - times[0]
    data = fh.variables['hgt'][:].data
    time = datetime.datetime.now().strftime('%H:%M:%S')
    return (year, times, lats, lons, data)

def get_data(hgt_path, yr_range = year_range(hgt_path)):
    yearly_data = []
    for year in range(yr_range[0], yr_range[1]+1):
        print(year)
        year, times, lats, lons, data = retrieve_data(os.path.join(hgt_path, f'hgt.{year}.nc'))
        yearly_data.append(data)
    return yearly_data
```
